### We gotta clean it up just for the columns 'Use' and 'Type of economic activity'

```python
def extract_english(VALUE):
    return VALUE.split('/')[-1]
```

```python
def translate_use(VALUE):
    if VALUE == 'Domestic':
        return 1
    if VALUE == 'Commercial':
        return 2
    if VALUE == 'Industrial':
        return 3
```

```python
df['Use'] = df['Use'].apply(extract_english)
df['Use'] = df['Use'].apply(translate_use)
```

```python
df['Use'].value_counts() 
```

```python
df['Type of economic activity'] = df['Type of economic activity'].apply(extract_english)
```

```python
# Ensure 'Date' is a datetime object
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Resample the data to daily frequency (if not already)
df_daily = df.resample('D').sum()

# Split data into training and test sets
train_size = int(0.8 * len(df_daily))
train, test = df_daily[:train_size], df_daily[train_size:]

# Create lag features
lag = 7  # Example: use 7 days of lag features
for i in range(1, lag + 1):
    train[f'lag_{i}'] = train['Consumption (L/h)'].shift(i)
    test[f'lag_{i}'] = test['Consumption (L/h)'].shift(i)

# Drop missing values introduced by lag features
train.dropna(inplace=True)
test.dropna(inplace=True)

# Define the target variable
y_train = train['Consumption (L/h)']
y_test = test['Consumption (L/h)']

# Define the features
X_train = train.drop(['Consumption (L/h)'], axis=1)
X_test = test.drop(['Consumption (L/h)'], axis=1)

# Initialize and train the XGBoost model
model = xgb.XGBRegressor()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate the Root Mean Squared Error (RMSE) as a performance metric
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```

```python
# Assuming 'y_test' contains the actual test set values and 'y_pred' contains the predicted values
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Actual', color='blue', marker='o')
plt.plot(y_test.index, y_pred, label='Predicted', color='red', marker='x')
plt.xlabel('Date')
plt.ylabel('Consumption (L/h)')
plt.title('Actual vs. Predicted Demand')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Assuming 'Date' is the datetime index and 'Consumption (L/h)' is the consumption column
# You can also add additional features and preprocessing as needed

# Calculate the z-score for the 'Consumption (L/h)' column
mean = df['Consumption (L/h)'].mean()
std = df['Consumption (L/h)'].std()
threshold = 2  # You can adjust the threshold based on your data

df['z_score'] = (df['Consumption (L/h)'] - mean) / std

# Identify anomalies based on the z-score threshold
anomalies = df[df['z_score'].abs() > threshold]
```

```python

# Based on the Elbow Method, select the optimal number of clusters
n_clusters = 3  # Adjust based on the visual analysis of the elbow plot

# Apply K-Means clustering
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(scaled_data)

# Add cluster labels to the original DataFrame
df['Cluster'] = kmeans.labels_

# Reduce dimensionality for visualization using PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
df['PC1'] = principal_components[:, 0]
df['PC2'] = principal_components[:, 1]

# Visualize customer segmentation
plt.figure(figsize=(10, 6))
plt.scatter(df['PC1'], df['PC2'], c=df['Cluster'], cmap='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Determine the optimal number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python

# Evaluate the clustering using silhouette score
silhouette_avg = silhouette_score(scaled_data, df['Cluster'])
print(f'Silhouette Score: {silhouette_avg}')

# Explore the characteristics of each customer segment
segmented_data = df.groupby('Cluster').mean()
print(segmented_data)
```
